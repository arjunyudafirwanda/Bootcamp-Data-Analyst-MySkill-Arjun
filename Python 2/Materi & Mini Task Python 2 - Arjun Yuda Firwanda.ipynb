# Indexing and Selection (slicing)
Select Columns:

Ada dua cara untuk memfilter nama kolom:

Metode 1: Cepat dan mudah, tetapi hanya berfungsi jika nama kolom TIDAK memiliki spasi atau karakter khusus.

    df.column_name               # returns series

Metode 2: Lebih kuat, dan dapat memfilter beberapa kolom.

    df['column']                  # returns series
    df[['column 1', 'column 2']]  # returns dataframe

Select Row:
Ada 2 cara utama untuk memilih baris:

    df.loc[label/value]    # filters by the labels/value of the index/column
    df.iloc[index]   # filters by the positions of the index/column

# Filtering Based on a Criteria
Untuk memfilter kerangka data berdasarkan suatu kondisi, kami cukup meneruskan kondisi tersebut sebagai vektor boolean.

Sebagai contoh, Mari memfilter kerangka data untuk menampilkan data di negara-negara Asia (AreaName = Asia).

```python
# 1. Create the condition boolean series
condition = df['Continent'] != 'Europe'
print(condition)
```

```python
# 2. Pass this condition into the dataFrame
df_non_europe = df[condition]
df_non_europe.head()
```

```python
# Filtering AreaNAme = Asia and RegName = Southern Asia
condition_1 = df['Continent']=='Asia'
condition_2 = df['Region Number']=='Southern Asia'
df[(condition_1) & (condition_2)]
```

# Mini Task
We can easily add more countries to line plot to make meaningful comparisons immigration from different countries.

Question: Let's compare the number of immigrants from India and China from 1980 to 2013.

```python
# Langkah 1: Dapatkan kumpulan data untuk China dan India, dan tampilkan kerangka data.
df_CI = df.loc[['China', 'India'], range(1980,2014)]
df_CI
```

```python
# Langkah 2: Buat data menjadi tranpose.
df_CI = df_CI.transpose()
df_CI.head()
```

```python
# Langkah 3: Plot grafik. Kami akan secara eksplisit menentukan plot garis dengan meneruskan parameter kind ke plot().
import matplotlib.pyplot as plt

df_CI.plot(kind='line')
plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
