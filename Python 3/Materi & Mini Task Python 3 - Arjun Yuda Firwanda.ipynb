```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load Dataset with URL

```python
# Load Dataset
URL = 'https://raw.githubusercontent.com/dataskillsboost/FinalProjectDA11/main/tokopaedi.csv'
df = pd.read_csv(URL)
df.head()
```

```python
# Fungsi info untuk menampilkan informasi data
df.info()
```

# Pengelolaan Data Date

```python
# Mengubah tipe kolom Date menjadi Datetime
df['order_date']= pd.to_datetime(df['order_date'])
df['ship_date']= pd.to_datetime(df['ship_date'])
df.dtypes
```

```python
# Membuat kolom tambahan untuk day, month, month_num (order_date)
df['day']=df['order_date'].dt.day_name()
df['month']=df['order_date'].dt.month_name()
df['month_num']=df['order_date'].dt.month
df['year']=df['order_date'].dt.year
df.head()
```

# Latihan Filter 1: Filter berdasarkan bulan October tahun 2015 (&)

```python
# Filter data
bulan = df['month'] == 'October'
tahun = df['year'] == 2015
df_filter2 = df[(bulan) & (tahun)]
df_filter2.head()
```

# Latihan Filter 2: Filter periode 15-30 Oktober 2015 untuk city Los Angeles

```python
# Cara 1
df.loc[(df['order_date'].dt.year == 2015) & 
       (df['order_date'].dt.month == 10) & 
       (df['order_date'].dt.day >= 15) & 
       (df['order_date'].dt.day <= 30)&
       (df['city'] == 'Los Angeles')].head()
```

```python
# Cara 2
start_date = df['order_date'] >= '2015-10-15'
end_date = df['order_date'] <= '2015-10-30'
city = df['city'] == 'Los Angeles'
df_filter3 = df[start_date & end_date & city]
df_filter3
```

# Penggunaan sort_values
Penggunaan Short Values:
Formula: nama_variabel = nama_dataframe.sort_values(by=[nama_kolom_1,nama_kolom_2],ascending=[urutan_1 , urutan_2])

```python
# Short values order_date dan quantity
df_sort = df.sort_values(by=['order_date','quantity'],ascending=[False,False]) #filtering
df_sort[['order_date','quantity']].head(5) #selecting
```

# Latihan Short Values 
Urutkan berdasarkan ship_date dari paling awal dan category secara alfabet. Tampilkan untuk 10 data paling atas, dengan menampilkan kolom: country,category,ship_date.

```python
df_sort1 = df.sort_values(by=['ship_date','category'],ascending=[True,True]) #filtering
# df_sort1
df_sort1[['country','category','ship_date']].head(10)
```

# Penggunaan fungsi pivot_table & groupby

# Fungsi pivot_table
Formula: nama_variable = pd.pivot_table(nama_dataframe, values=['kolom_nilai'],index=['baris'],columns=['header'],aggfunc=np.jenis_agregasi).

Output: DataFrame

```python
# Pivot table dengan values quantity, index region, columns category
df_pivot = pd.pivot_table(df,values=['quantity'],index=['region'],columns=['category'],aggfunc=np.sum)
df_pivot
```

# Latihan Pivot Table 1 
Latihan: Buatlah tabel pivot untuk: index = country, header/kolom = city, nilainya = penjumlahan dari profit

```python
df_pivot1 = pd.pivot_table(df,values=['profit'],index=['country'],columns=['city'],aggfunc=np.sum)
df_pivot1
```

# Latihan Pivot Table 2
Buatlah tabel pivot untuk index = segmen, value = profit, kolom = city, agregasi = rata-rata

```python
df_pivot2 = pd.pivot_table(df,values=['profit'],index=['segment'],columns=['city'],aggfunc=np.average)
df_pivot2
```

# Penggunaan Groupby
Formula: nama variabel = nama_dataframe.groupby(by=['kolom_1' , 'kolom_2'])['kolom_agregasi'].jenis_agregasi().

Output: series

```python
# Fungsi groupby mount_num month dan profit
df_groupby = df.groupby(by=["month_num","month"])["profit"].sum()
df_groupby
```

```python
# Merubah output groupby menjadi DataFrame
df_groupby_new = pd.DataFrame(df_groupby.reset_index(name='Total')) #reset_index untuk memberi header pada kolom hasil groupby
df_groupby_new
```

# Latihan Groupby
Buatlah dengan fungsi groupby, top 5 city penghasil profit paling besar selama periode 2015.

```python
# Cara 1
year = df['year']==2015 #condition
df_filter4 = df[year] # filter
df_groupby1 = df_filter4.groupby(by=["city"])["profit"].sum() #group by
df_groupby2 = pd.DataFrame(df_groupby1.reset_index(name='profit')) #kita ubah si group by menjadi dataframe
df_sort2 = df_groupby2.sort_values(by=['profit'],ascending=[False]) #sorting
df_sort2.head(5) # menampilkan top 5
```

```python
# # Cara 2
df_filter5 = df[df['year'] == 2015]
df_groupby3 = df_filter5.groupby(by=["city"])["profit"].sum()
df_groupby4 = pd.DataFrame(df_groupby3.reset_index(name='Total'))
df_groupby4.sort_values(by='Total', ascending=False).head()
```

# Penggunaan Fungsi Dataframe
Membuat dan menggunakan fungsi sederhana pada DataFrame.
Formula: def nama_fungsi (value):

```python
#Membuat suatu fungsi
def status_qty(value):
    if value > 2:
      return 'High'
    elif value == 2:
      return 'Medium'
    else:
        return 'Low'
```

```python
# Pengaplikasian fungsi
df['status_qty'] = df['quantity'].apply(status_qty)
df.head()
```

# Latihan Fungsi DataFrame
Buatlah suatu kolom baru yang bernama ship_delay yang dapat menyatakan bahwa apabila waktu pengiriman <= 4 hari dari tanggal order maka 'Good' selain itu maka 'Bad'

```python
#Membuat suatu fungsi
def condition_test(y):
    if y <= 4:
      return 'Good'
    else:
      return 'Bad'
```

```python
# Pengaplikasian fungsi
df['diff_delay'] = df['ship_date'] - df['order_date']
df['diff_delay_num'] = df['diff_delay'].apply(lambda x: int(x.total_seconds() / (24 * 60 * 60)))
df['ship_delay'] = df['diff_delay_num'].apply(condition_test)
df.head()
```

# Penggunaan SQL in Tools Python (Jupyter Notebook)

```python
# Menjalankan SQL
from sqlite3 import connect
conn = connect(':memory:')
df.to_sql('tokopaedi',conn, index=False, if_exists='replace')
```

```python
# Query SQL
df_sql = pd.read_sql("""
SELECT
    tokopaedi.city,
    sum(tokopaedi.profit) profit
from
    tokopaedi
group by
    1
order by
    2 desc
limit 10
""", conn)
df_sql
```

```python
# Membuat visualisasi sql
df_sql.plot(x='city',
           y=['profit'],
           kind='bar',
           grid = True,
           xlabel = 'City',
           ylabel = 'Profit',
           figsize=(12,7),
           rot = 0, #Rotasi
           title = 'Monthly Total Profit',
           table = False,
           sort_columns = False,
           secondary_y = False)
```
